```python
# NAO ALTERE OS COMANDOS ABAIXO
import gdown
import pandas as pd

url = 'https://drive.google.com/uc?id=1_DdGJ4yOEKhaER-dabRtgK5qp3vP2HiHFO5qus9hHRs'
output = 'altura_peso.xlsx'
gdown.download(url, output, quiet=True)

dados = pd.read_excel('/content/altura_peso.xlsx')
```

>
> Ao analizar os dados de peso e altura, a variável `sexo` foi totalmente desconsiderada. E isto pode levar à uma má compreensão dos dados pelo algoritmo.
>
> É bem conhecido que, em média, homens são mais altos e mais pesados do que mulheres. Será que os dados nos apresentam isso?

###Questão 1

Crie 2 conjuntos de dados: um contendo só as informações de homens e o outro somente de mulheres.

```python
# Questao 1
dados_homens=dados[dados['Sexo']=='Masculino']
dados_mulheres=dados[dados['Sexo']=='Feminino']
display(dados_homens,dados_mulheres)
```

###Questão 2

Calcule a média de altura e peso para homens e mulheres. Qual é a diferença absoluta entre elas?

```python
# Questao 2
Media_Altura_Homens=dados_homens['Altura'].mean()
Media_Peso_Homens=dados_homens['Peso'].mean()
Media_Altura_Mulheres=dados_mulheres['Altura'].mean()
Media_Peso_Mulheres=dados_mulheres['Peso'].mean()
#caculando a média absoluta entre homens e mulheres
MediaAbsoluta_Altura=(max(Media_Altura_Homens,Media_Altura_Mulheres)-min(Media_Altura_Homens,Media_Altura_Mulheres))
MediaAbsoluta_Peso=(max(Media_Peso_Mulheres,Media_Peso_Homens)-min(Media_Peso_Homens,Media_Peso_Mulheres))
print(f'Media de altura entre os homens {Media_Altura_Homens:.2f}cm \nMedia de peso entre os homens {Media_Peso_Homens:.2f}kg \nMedia de altura entre as mulheres {Media_Altura_Mulheres:.2f}cm \nMedia de peso entre as mulheres {Media_Peso_Mulheres:.2f}kg \n')
print(f'A diferença absoluta de peso e altura entre homens e mulheres é {MediaAbsoluta_Peso:.2f}kg e {MediaAbsoluta_Altura:.2f}cm (respectivamente)')
```

###Questão 3

Calcule a mediana de altura e peso para homens e mulheres. Qual é a diferença absoluta entre elas?

```python
# Questao 3
Mediana_Altura_Homens=dados_homens['Altura'].median()
Mediana_Peso_Homens=dados_homens['Peso'].median()
Mediana_Altura_Mulheres=dados_mulheres['Altura'].median()
Mediana_Peso_Mulheres=dados_mulheres['Peso'].median()
print(f'Mediana peso homens:{Mediana_Peso_Homens}kg')
print(f'Mediana altura homens:{Mediana_Altura_Homens}cm')
print(f'Mediana peso mulheres:{Mediana_Peso_Mulheres}kg')
print(f'Mediana altura homens:{Mediana_Altura_Mulheres}cm')
MediaAbsoluta_MedianaAltura=(max(Mediana_Altura_Homens,Mediana_Altura_Mulheres)-min(Mediana_Altura_Homens,Mediana_Altura_Mulheres))
MediaAbsoluta_MedianaPeso=(max(Mediana_Peso_Homens,Mediana_Peso_Mulheres)-min(Mediana_Peso_Homens,Mediana_Peso_Mulheres))
print(f'\nA diferença absoluta das medianas de peso e altura entre homens e mulheres é {MediaAbsoluta_MedianaPeso:.2f}kg e {MediaAbsoluta_MedianaAltura:.2f}cm (respectivamente)')
```

###Questão 4

Calcule o desvio-padrão de altura e peso para homens e mulheres.

> Responda: Necessariamente, quem tem maior média, tem maior desvio-padrão?

```python
# Questao 4
dp_Peso_Homens= dados_homens['Peso'].std()
dp_Altura_Homens=dados_homens['Altura'].std()
dp_Peso_Mulheres= dados_homens['Peso'].std()
dp_Altura_Mulheres=dados_mulheres['Altura'].std()
print(f'Desvio Padrão homens \n Peso: {dp_Peso_Homens:.2f}kg\n Altura: {dp_Altura_Homens:.2f}cm\n')
print(f'Desvio Padrão mulheres \n Peso: {dp_Peso_Mulheres:.2f}kg\n Altura: {dp_Altura_Mulheres:.2f}cm')
#Respondendo a pergunta "Necessariamente, quem tem maior média, tem maior desvio-padrão?"
#A resposta é não pois a média calcula o valor típico daqueles dados enquanto o desvio padrão mede a distribuição dos dados(disperção), com isso é possível que um grupo tenha uma média alta e um desvio padrão pequeno
```
