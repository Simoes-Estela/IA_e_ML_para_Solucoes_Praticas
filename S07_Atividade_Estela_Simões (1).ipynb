###Questão 1

Calcule a média salarial de homens e mulheres.

```python
# Questao 1
salarios_homens=salarios[salarios['Sexo']=='Masculino']
salarios_mulheres=salarios[salarios['Sexo']=='Feminino']
Media_Salarial_homens=salarios_homens['Salario'].mean()
Media_Salarial_mulheres=salarios_mulheres['Salario'].mean()
print(f'A média salárial entre homens é de R${Media_Salarial_homens:.2f} enquando a de mulheres é R${Media_Salarial_mulheres:.2f}')
```

###Questão 2

Em média, qual sexo possui mais anos de experiência?

```python
# Questao 2
#para isso vou criar 2 váriaveis com a media de anos de experiência entre homens e mulheres
Media_Experiência_Homens=salarios_homens['Anos_Experiencia'].mean()
Media_Experiência_Mulheres=salarios_mulheres['Anos_Experiencia'].mean()
print(f'A média de experiência entre homens é {Media_Experiência_Homens:.2f} anos a de mulheres é de {Media_Experiência_Mulheres:.2f} anos')
if(Media_Experiência_Mulheres>Media_Experiência_Homens):
  print('O SEXO FEMININO POSSUI MAIS ANOS DE EXPERIÊNCIA')
else:
    print('O SEXO MASCULINO POSSUI MAIS ANOS DE EXPERIÊNCIA')

```

###Questão 3

Existe associação entre `idade`, `anos de experiência` e `salario`? Compute a matriz de correlação. Os valores mudam considerando o `sexo`?

```python
# Questao 3
Matriz_correlação=salarios[['Anos_Experiencia','Salario','Idade']].corr() #Matriz de correlação dos salários de TODOS
Matriz_correlação_homens=salarios_homens[['Anos_Experiencia','Salario','Idade']].corr() #Matriz de correlação dos salários MASCULINA
Matriz_correlação_mulheres=salarios_mulheres[['Anos_Experiencia','Salario','Idade']].corr() #Matriz de correlação dos salários FEMININA
display(Matriz_correlação,Matriz_correlação_mulheres,Matriz_correlação_homens)
#Há correlação entre a idade, anos de experiência e salário quem tem mais experiência possui um salario maior comparado há idade
#Os valores mudam considerando o sexo.Homens possuem um salário maior considerando suas idades e anos de experiência do que as mulheres
```
