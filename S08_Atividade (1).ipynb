```python
# NAO ALTERE OS COMANDOS ABAIXO
import gdown
import pandas as pd

url = 'https://drive.google.com/uc?id=1DwqWeEuzjuMn_WAmz7U3Uo49I1zeaZyg'
output = 'MFGEmployees4.xlsx'
gdown.download(url, output, quiet=True)

dados_funcionarios = pd.read_excel('/content/MFGEmployees4.xlsx')
```

###Questão 1

Verifique se há registros duplicados na base. Caso positivo, remova-os.

```python
# Questao 1
#Criei uma estrutura de decisão, caso haja duplicados na base ele realiza a ação de removelos

if( (dados_funcionarios.duplicated().any()) == True):
  print(f"há {dados_funcionarios.duplicated().sum()} dados duplicados na base")
  print("Criando uma nova base sem os dados duplicados...")
  df_dados_func = dados_funcionarios.drop_duplicates()
  print(f"Agora há {df_dados_func.duplicated().sum()} dados duplicados na base")
else:
  print("Não há dados duplicados na base")
```

###Questão 2

Verifique se há valores ausentes na base. Caso positivo, remova-os.

```python
# Questao 2
#utilizei a mesma técnica da questão 1

if((df_dados_func.isna().any()).any() == True):
    print(f"Há {df_dados_func.isna().sum().sum()} valores ausentes na base")
    print("Criando uma nova base sem os valores ausentes...")
    df_dados_func = df_dados_func.dropna()
    print(f"agora há {df_dados_func.isna().sum().sum()} valores ausentes na base")
else:
  print("Não há valores ausentes na base")

```

###Questão 3

Há variáveis que funcionam como "rótulos" nesta base. Quais são? Remova-as

```python
# Questao 3
df_dados_func.info()
#Sim, há três variáveis que funcionam como rótulos nesta base: sobrenome,nome e id
df_dados_func = df_dados_func.drop(['id','sobrenome','nome'], axis=1) #removendo as 3 variáveis
df_dados_func.info()
```

###Questão 4

Existem valores suspeitos ou discrepantes para alguma variável? Caso positivo, remova-os.

Por lei:
*     A idade mínima de um funcionário é 14 anos e máxima é de 80 anos.
*     A duração do serviço não pode ultrapassar 44 horas.

```python
# Questao 4
df_dados_func.describe()
# há valores suspeitos e discrepantes nas variáveis de idade e duração de serviço
```

```python
#continuação da Questao 4
#filtrando todos os valores em que: a  IDADE seja maior que 14 anos e menor que 80 anos, e a DURAÇÃO DE SERVIÇO menor que  44 horas
df_dados_func = df_dados_func[(df_dados_func['idade']>14) &
                              (df_dados_func['idade']<80) &
                              (df_dados_func['duracao_servico']<=44)]
df_dados_func.describe()
```
