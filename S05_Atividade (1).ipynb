```python
# NAO ALTERE OS COMANDOS ABAIXO
import gdown
import pandas as pd

url = 'https://drive.google.com/uc?id=1ogSBybCNgijjL1tVksCzwzPzf7r3cY0T'
output = 'pacientes_cardiacos.xlsx'
gdown.download(url, output, quiet=True)

pacientes_cardiacos = pd.read_excel('/content/pacientes_cardiacos.xlsx')
```

Este conjunto de dados médicos é utilizado para prever a presença ou ausência de doenças cardíacas em pacientes, com base em uma série de atributos clínicos. Ele foi coletado por várias instituições médicas e inclui informações sobre características demográficas, exames clínicos e resultados de testes relacionados à saúde cardíaca.

O conjunto de dados contém 14 atributos, que incluem:

*   `id`: Identificador único de cada paciente.
*   `age`: Idade do paciente (em anos).
*   `sex`: Sexo do paciente (Masculino/Feminino).
*   `cp`: Tipo de dor no peito (angina típica, angina atípica, não-anginal, assintomático).
*   `trestbps`: Pressão arterial de repouso (em mm Hg no momento da admissão).
*   `chol`: Colesterol sérico (em mg/dL).
*   `fbs`: Se o nível de glicose no sangue em jejum é superior a 120 mg/dL.
*   `restecg`: Resultados do eletrocardiograma de repouso (normal, anormalidade STT, hipertrofia ventricular esquerda).
*   `thalach`: Frequência cardíaca máxima alcançada.
*   `exang`: Angina induzida por exercício (Verdadeiro/Falso).
*   `oldpeak`: Depressão do segmento ST induzida por exercício, em comparação ao descanso.
*   `slope`: A inclinação do segmento ST no pico do exercício.
*   `ca`: Número de vasos principais (0-3) coloridos por fluoroscopia.
*   `thal`: Defeito fixo, defeito reversível ou normal.
*   `num`: A variável predita, que indica a presença ou ausência de doença cardíaca (geralmente 0 ou 1).

O objetivo principal deste dataset é prever se um paciente tem ou não doença cardíaca com base nos atributos fornecidos. Além disso, ele oferece uma excelente oportunidade para diagnóstico experimental e extração de insights que podem ajudar na compreensão dos problemas relacionados a doenças cardíacas.

###Questão 2

Dado que a coluna `id` é apenas um identificador único, remova essa coluna do dataset e salve o resultado em um objeto chamado `df_pacientes`.

```python
# Questao 2
df_pacientes=pacientes_cardiacos.drop(['id'],axis=1)
df_pacientes
```

###Questão 3
